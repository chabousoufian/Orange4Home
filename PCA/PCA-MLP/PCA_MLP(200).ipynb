```python
X = data[training_labels]
y = data[classes_labels]
y = y.values
X = X.values
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=200)
```

```python
pca.fit(X)
```

```python
x_pca=pca.transform(X)
```

```python
X.shape
```

```python
x_pca.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca,y, test_size = 0.3)
x_test, x_val, y_test, y_val = train_test_split(x_test,y_test, test_size = 0.3)
```

```python
# hyper-parameters
num_classes = 25


model = Sequential()
# first layer with 
model.add(Dense(x_train.shape[1], activation='relu', input_shape=(x_train.shape[1],)))
# second layer with 
model.add(Dense((x_train.shape[1]+num_classes)/2, activation='relu'))
# Dropout of 50%
model.add(Dropout(0.5))
# Camada de 20 neurônios
model.add(Dense((x_train.shape[1]+num_classes)/2, activation='relu'))

# Classification Layer 
model.add(Dense(num_classes, activation='softmax'))

model.summary()


sgd = SGD(learning_rate=0.001)
model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])


batch_size = 34
epochs = 30
history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_val, y_val))
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,8))
ax[0].plot(history.history['loss'], color='b', label="Training loss")
ax[0].plot(history.history['val_loss'], color='r', label="validation loss",axes =ax[0])
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history.history['accuracy'], color='b', label="Training accuracy")
ax[1].plot(history.history['val_accuracy'], color='r',label="Validation accuracy")
legend = ax[1].legend(loc='best', shadow=True)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

score = model.evaluate(x_val, y_val, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
